```python
from graphviz import Digraph

def visualize_tree(tree):
    if tree is None: return 'Nothing in the tree!'
    def add_nodes_edges(tree, dot=None):
        if dot is None:
            dot = Digraph()
            dot.attr('node', shape='circle')
            dot.node(name=str(tree), label=str(tree.val))

        for child in [tree.left, tree.right]:
                if child is not None:
                    if child == tree.left: dot.attr('node', shape='circle', style='filled', color='lightblue')
                    if child == tree.right: dot.attr('node', shape='circle',style='filled', color='lightgreen')
                    dot.node(name=str(child) ,label=str(child.val))
                    dot.edge(str(tree), str(child))
                    dot = add_nodes_edges(child, dot=dot)
        return dot

    dot = add_nodes_edges(tree)
    display(dot)
    
```

```python
class TreeNode:
    def __init__(self,x):
            self.val = x
            self.left = None
            self.right = None
```

```python
t1 = TreeNode(1)
t1.left = TreeNode(3)
t1.right = TreeNode(2)
visualize_tree(t1)
```

```python
t1.left.left = TreeNode(5)
t1.left.right = TreeNode(15)

t1.right.right = TreeNode(55)
t1.right.right.left = TreeNode(86)
t1.right.right.right = TreeNode(101)
t1.right.right.right.right = TreeNode(1019)
t1.right.right.right.left = TreeNode(200)
visualize_tree(t1)
```

```python
def dfs(self):
    print(self.val)

    if self.left:
        self.left.dfs()
    if self.right:
        self.right.dfs()

TreeNode.dfs = dfs
```

```python
t1.dfs()
```

```python
def dfs_inorder(self):
    if self.left:
        self.left.dfs_inorder()
    print(self.val)
    if self.right:
        self.right.dfs_inorder()

TreeNode.dfs_inorder = dfs_inorder
```

```python
t1.dfs_inorder()
```

```python
def dfs_postorder(self):
    if self.left:
        self.left.dfs_postorder()
    if self.right:
        self.right.dfs_postorder()
    print(self.val)

TreeNode.dfs_postorder = dfs_postorder
```

```python
t1.dfs_postorder()
```

```python
def bfs(self):
    to_visit = [ self ]
    while to_visit:
        current = to_visit.pop(0)
        print(current.val)
        if current.left:
            to_visit.append(current.left)
        if current.right:
            to_visit.append(current.right)

TreeNode.bfs = bfs
```

```python
t1.bfs()
```

```python
def dfs_apply(self,fn):
    fn(self)
    if self.left:
        self.left.dfs_apply(fn)
    if self.right:
        self.right.dfs_apply(fn)

TreeNode.dfs_apply = dfs_apply
```

```python
class PerformSum:
    def __init__(self):
        self.sum = 0

    def process(self,node):
        self.sum += node.val

    def get_sum(self):
        return self.sum

    def reset_sum(self):
        self.sum = 0

p = PerformSum()
```

```python
t1.dfs_apply(p.process)
print(p.get_sum())
p.reset_sum()
```
