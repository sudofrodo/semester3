```python
def heapify(lst , n, root):
    largest = root
    l = 2*root + 1
    r = 2*root + 2

    if l < n and lst[l] > lst[largest]:
        largest = l

    if r < n and lst[r] > lst[largest]:
        largest = r

    if largest != root:
        lst[root], lst[largest] = lst[largest], lst[root]

        heapify(lst, n, largest)
    
    
```

```python
def build_heap(lst):
    n = len(lst)
    for i in reversed( range(n // 2)):
        heapify(lst,n,i)
```

```python
heap = [100,5,3,2,8,15,6,102]
```

```python
print(heap)
build_heap(heap)
print(heap)
```
