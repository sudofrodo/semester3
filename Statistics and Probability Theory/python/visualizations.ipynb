```python
url = "data/nhanes_2015_2016.csv"
da = pd.read_csv(url)
```

```python
gender = da['RIAGENDR']
gender.value_counts()
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gender.value_counts().plot(kind='bar')
plt.show()
```

```python
gender.value_counts().plot(kind='bar')
plt.xticks([0,1] , ['Male', 'Female'])
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```

## Most Common Values

```python
sizes = np.array([22,34,56,75,68,46,89,43,45,24,23,11,45])
```

```python
sizes.mean()
```

```python
from collections import Counter
cnt = Counter()
for size in sizes:
    cnt[size] += 1
cnt.most_common()
```
